```python
from datetime import timedelta
import numpy as np
import pandas as pd
import xarray as xr
import cartopy.crs as ccrs 
import matplotlib.pyplot as plt
```

```python
ds0 = xr.open_dataset('/glade/derecho/scratch/molina/ml_coordsv2.nc')
```

```python
lat_vals = [45, 25, 24, -23, -55, 22, -12, -44, 35, 5, 0, -35]
lon_vals = [190, 230, 262, 280, 280, 112, 110, 125, 0, 60, 0, 10]
```

```python
plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.PlateCarree())

for lat_val, lon_val in zip(lat_vals, lon_vals):
    
    tmp = ds0.sel(x=slice(lon_val, lon_val+31), y=slice(lat_val, lat_val+31))
    ax.pcolormesh(tmp.x.values, tmp.y.values, tmp.top.transpose('y','x').values, 
                  shading='gouraud', transform=ccrs.PlateCarree())

ax.coastlines()
ax.set_extent([-180,180,-90,90], ccrs.PlateCarree())

plt.show()
```
