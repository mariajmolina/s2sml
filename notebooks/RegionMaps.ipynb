## Plotting regions

```python
import matplotlib.pyplot as plt
import numpy as np
from cartopy import crs as ccrs, feature as cfeature
import matplotlib.patches as mpatches
```

```python
fig = plt.figure(figsize=(11, 8))

ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_global() # this line is key when you don't have underlying data
ax.coastlines()
ax.gridlines(draw_labels=True)


# Southern Hemisphere region
ax.add_patch(mpatches.Rectangle(xy=[-80, -55], # location of bottom-left corner
                                width=32, height=32, # width/height in degrees
                                facecolor='blue',
                                alpha=0.2,
                                transform=ccrs.PlateCarree()))

# Northern Hemisphere region
ax.add_patch(mpatches.Rectangle(xy=[-120, 15],
                                width=32, height=32,
                                facecolor='red',
                                alpha=0.2,
                                transform=ccrs.PlateCarree()));

plt.savefig('../figures/regions.png')
```
