# Testing for sources of randomness

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Read trial results

```python
results_dir_1 = 'randtest_622/' # set results directory
case1 = 'trial_results_1233269.desched1.csv'

results_dir_2 = 'randtest_622_v2/'
case2 = 'trial_results_1233271.desched1.csv'

df_1 = pd.read_csv("../"+results_dir_1+"trial_results/"+case1)
df_2 = pd.read_csv("../"+results_dir_2+"trial_results/"+case2)
```

```python
df_1
```

```python
df_2
```

```python
plt.hist(df_1.learning_rate, bins=5, label='test 1')
plt.hist(df_2.learning_rate, bins=5, alpha=0.7, label='test 2')
plt.xlabel('Learning Rate')
plt.ylabel('Counts')
plt.legend();

#plt.savefig('../figures/randtest_622_lr.png')
```

```python
plt.hist(df_1.train_batch_size, bins=20, label='test 1')
plt.hist(df_2.train_batch_size, bins=20, alpha=0.7, label='test 2')
plt.xlabel('Batch size')
plt.ylabel('Counts')
plt.legend();

#plt.savefig('../figures/randtest_622_batch.png')
```
